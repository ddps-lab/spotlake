```python
from parse import *
import pandas as pd
from IPython.display import display
import json
import os
```

```python
def searchJsonFile(path):
    filenames = os.listdir(path)
    for filename in filenames:
        full_name = os.path.join(path, filename)
        ext = os.path.splitext(full_name)[-1]
        if ext != '.json':
            filenames.remove(filename)
    return filenames
```

```python
def decodeJson(data_table, json_data_list):
    # decode json data (dict)
    for data in json_data_list :
        # extract update information for each message
        last_updated = data["last_updated"]
        latest_updated = data["latest_updated"]
        outer_skus_list = data["skus"]
        
        for oskus in outer_skus_list :
            # extract category, cloud_service_id for each outer skus
            category = oskus["category"]
            cloud_service_id = oskus["cloud_service_id"]
            inner_skus_list = oskus["skus"]
            
            for inskus in inner_skus_list :
                # extract description, service_region, prices, sku_id
                description = inskus['description']
                service_regions = inskus['service_regions']
                prices = inskus['prices']
                sku_id = inskus['sku_id']
                
                # add to dataframe
                new_df = pd.DataFrame(data=[[category, cloud_service_id, description, sku_id, prices, service_regions, last_updated, latest_updated]], columns=col_list)
                if data_table.empty == True :
                    data_table = new_df
                else:
                    data_table = pd.concat([data_table, new_df])
    
    return data_table
                
```

```python
if __name__ == '__main__':
    # search raw data json files
    path = '../../../gcp_raw_data/'
    file_list = searchJsonFile(path)
    
    global col_list
    col_list = ['category', 'cloud_service_id', 'description', 'sku_id',
                'prices', 'service_regions', 'last_updated', 'latest_updated']
    
    for file in file_list :
        json_data_list = []
        # open json file
        with open(path + file, 'r', encoding='utf-8') as f:
            json_data_list = json.load(f)
        
        # create Dataframe
        data_table = pd.DataFrame(columns=col_list)
        
        # decode json to dataframe
        result_data_table = decodeJson(data_table, json_data_list)
        
        display(result_data_table)
```
